### Basic calculation of profit
- Extract the lien interest paid for each property. This not easy and we have probably done it wrong.
- Identify when the lien was paid off.
- This takes a long time to run.

````python
Results = pd.read_pickle('All_Results.pkl')
ag = pd.DataFrame(columns=['Parcel_ID', 'Tax_Year', 'Lien_Interest_Charge', 'Lien_Interest_Payment'])
for pid in Results.Parcel_ID:
    for y in Results.Tax_Year[Results.Parcel_ID == pid]:
        c = Account_Details.Amount[(Account_Details.Tax_Type == "lien interest charge") & 
                           (Account_Details.Tax_Year == y) & (Account_Details.Parcel_ID == pid)].sum()
        p = Account_Details.Amount[(Account_Details.Tax_Type == 'lien interest payment') & 
                           (Account_Details.Tax_Year == y) & (Account_Details.Parcel_ID == pid)].sum()
        ec = Account_Details.Amount[(Account_Details.Tax_Type == "lien endorsement") & 
                           (Account_Details.Tax_Year == y) & (Account_Details.Parcel_ID == pid)].sum()
        ep = Account_Details.Amount[(Account_Details.Tax_Type == "lien endorsement") & 
                           (Account_Details.Tax_Year == y) & (Account_Details.Parcel_ID == pid)].sum()
        ag = ag.append({'Parcel_ID': pid, 'Tax_Year': y, 'Lien_Interest_Charge': c, 'Lien_Interest_Payment': p},
                       ignore_index=True)
        #print(pid, y, c, p)

Results = Results.merge(ag)
Results['First_Year_Profit'] = Results.Lien_Interest_Charge - Results.Winning_Bid
````
